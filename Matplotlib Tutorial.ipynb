```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

plt.bar(labels, values)
plt.show()
```

```python
# Adding Patterns bars

labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.show()
```

```python
# Adding Patterns bars with 'for loop'

labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

# adding title
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(gas.Year, gas.USA, label = 'USA')         # 1st way to plot a x and y axis from given dataset
plt.plot(gas['Year'], gas['Canada'], label = 'Canada')# 2nd way to plot a x and y axis from given dataset
plt.plot(gas['Year'], gas['South Korea'], label = 'South Korea') # Used in objects with 2 names eg. 'South Korea'

# Adding legend
plt.legend()

plt.show()
```

```python
# adding x and y ticks

gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

# adding title
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(gas.Year, gas.USA,'b.-', label = 'USA')                       
plt.plot(gas['Year'], gas['Canada'],'r.-', label = 'Canada')           
plt.plot(gas['Year'], gas['South Korea'],'g.-', label = 'South Korea') 

# x and y label

plt.xlabel('Years')
plt.ylabel('USD')

# Adding legend
plt.legend()

plt.xticks(gas.Year[::3]) # skips 3 years from the first to the last

#print(gas.Year)

plt.show()
```

```python
# using for loop to check all the countries

gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

# adding title
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=(country))

# x and y label

plt.xlabel('Years')
plt.ylabel('USD')

# Adding legend
plt.legend()

plt.xticks(gas.Year[::3]) # skips 3 years from the first to the last


plt.show()
```

```python
# Using for loop to check specific countries


gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

# adding title
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

countries_to_look_at = ['Australia', 'USA', 'Germany', 'UK', 'France']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.', label=(country))

# x and y label

plt.xlabel('Years')
plt.ylabel('US Dollars')

# Adding legend
plt.legend()

plt.xticks(gas.Year[::3]) # skips 3 years from the first to the last


plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5) # Displays the first 5 rows in the dataset
```

### Histogram

```python
# displaying a histogram 'plt.hist()'

bins = [40,50,60,70,80,90,100] #intervals of the bins/divisions in the histogram

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bins, color='#abcdef')  # color can be written in name form or hexadecimal form

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')

plt.title('Distribution of Player Skills in FIFA 2018')

plt.grid(linestyle="--", alpha=1) #'alfa' - the saturation of the grid line (0-1)

plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%') # 'autopct' - adds the percentage to the pie chart

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### Pie Chart #2

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot') # plotting style


fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4) # corresponds to the labels

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
# pctdistance - percentage distance of the %values from the center of the pie chart
# explode - Breaks the graph apart abit
plt.show()
```

```python
# More Customization

plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall'] # filter out the players in Barcelona-'Fc Barcelona' and display their overall score-'Overall'
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']     # filter out the players in 'Real Madrid' and display their overall score-'Overall'
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall'] # filter out the players in 'New England Revolution' and display their overall score-'Overall'

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))

bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})
# 'parch_artist=True' - Enables change of face color for the boxes
# 'medianprops={'linewidth': 2}' - Changes the size of the medium line

plt.title('Professional Soccer Team Comparison')
plt.xlabel('Teams')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
